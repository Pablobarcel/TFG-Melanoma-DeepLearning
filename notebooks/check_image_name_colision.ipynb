```python
from pathlib import Path
import pandas as pd

# --------------------------------------------------
# CONFIGURACIÓN
# --------------------------------------------------
ISIC_DIR = Path("../images_ISIC")
MALIGNANT_DIR = Path("../imagenes_Malignant")

OUTPUT_CSV = "image_name_collisions.csv"

VALID_EXTENSIONS = {".jpg", ".jpeg", ".png", ".bmp", ".tiff"}

# --------------------------------------------------
# FUNCIÓN PARA OBTENER NOMBRES DE IMÁGENES
# --------------------------------------------------
def get_image_names(folder: Path):
    return {
        p.name
        for p in folder.iterdir()
        if p.is_file() and p.suffix.lower() in VALID_EXTENSIONS
    }

# --------------------------------------------------
# MAIN
# --------------------------------------------------
def main():
    if not ISIC_DIR.exists():
        raise FileNotFoundError(f"No existe la carpeta: {ISIC_DIR}")
    if not MALIGNANT_DIR.exists():
        raise FileNotFoundError(f"No existe la carpeta: {MALIGNANT_DIR}")

    isic_names = get_image_names(ISIC_DIR)
    malignant_names = get_image_names(MALIGNANT_DIR)

    print(f"[INFO] Imágenes ISIC: {len(isic_names)}")
    print(f"[INFO] Imágenes Malignant: {len(malignant_names)}")

    # Colisiones exactas de nombre
    collisions = sorted(isic_names.intersection(malignant_names))

    if len(collisions) == 0:
        print("\n✅ NO se han detectado colisiones de nombres.")
        print("Puedes consolidar las imágenes manteniendo los nombres originales.")
    else:
        print(f"\n⚠️  SE HAN DETECTADO {len(collisions)} COLISIONES DE NOMBRES.")
        print("Ejemplos:")
        for name in collisions[:10]:
            print(f"  - {name}")

        # Guardar CSV para auditoría
        df = pd.DataFrame({
            "image_name": collisions,
            "in_ISIC": True,
            "in_Malignant": True
        })
        df.to_csv(OUTPUT_CSV, index=False)

        print(f"\n📄 Lista completa guardada en: {OUTPUT_CSV}")
        print("RECOMENDACIÓN: añadir prefijo por source_db al consolidar.")

if __name__ == "__main__":
    main()
```
